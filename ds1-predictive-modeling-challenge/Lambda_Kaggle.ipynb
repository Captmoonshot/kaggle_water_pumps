```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reat the data into pandas DataFrames

df = pd.read_csv('train_features.csv')
df_test = pd.read_csv('train_labels.csv')

df.head()
```

```python
df_test.head()
```

```python
df_test.shape
```

```python
# Combine the DataFrames 
# Remember, target variable is 'status_group'

df_combined = df.merge(df_test, on='id', how='left')
df_combined.head()
```

```python
df_combined.dtypes
```

```python
df_combined.isna().sum()
```

```python
# Just the numerical data

df_combined_num = df_combined.select_dtypes(include=['number']).copy()
df_combined_num.head()
```

```python
# add the target variable to the numeric

df_combined_num['target'] = df_combined['status_group']
df_combined_num.head()
```

```python
# no null values for the Numeric DataFrame

df_combined_num.isna().sum()
```

```python
df_combined_num.shape
```

```python
# target variable seems to be shuffled

df_combined_num['target'].head(100)
```

```python
df_combined_num['target'].value_counts(normalize=True)
```

#### The majority class is funtioncal (54%), so 54% is the baseline accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = df_combined_num.loc[:, 'id':'construction_year']
y = df_combined_num['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LogisticRegression().fit(X_train, y_train)

y_preds = lr.predict(X_test)

print("Training set score: {:.2f}".format(lr.score(X_train,y_train)))
print("Test set score: {:.2f}".format(lr.score(X_test, y_test)))
```
